```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.catplot(x="sex",y="age",data=titanic)
```

```python
sns.lineplot(x="survived",y="alive",data=titanic)
```

```python
sns.scatterplot(x="age",y="fare",data=titanic)
```

```python
sns.countplot(x="sex",data=titanic)
plt.title("Count of male & female",color='black',fontsize=20)
plt.show( )
```

```python
sns.histplot(titanic['fare'])
```

```python
sns.violinplot(x="sex",y="age",data=titanic)
```

```python
sns.barplot(x="survived",y="sibsp",data=titanic)
plt.figure(figsize=(10,6))
plt.show()
```

```python
x = titanic['pclass']
y = titanic['sibsp']
plt.plot(x,y,color='blue')
plt.show()
```

```python
x = titanic['survived']
y = titanic['sibsp']
z = titanic['parch']
plt.bar(x,y,z)
plt.show()
```

```python
sns.boxplot(x="age",y="sex",data=titanic)
plt.title("Boxplot of male & female",color='pink')
plt.show()
```
